```
#import require Library
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```
#simple neural network, has 1 layer that layer has 1 neuron. input shape just 1 value
#Keras(Deep learning python library) is a high-level neural networks API
#Keras is a model, a way to organize layers. 
#Sequential model, a linear stack of layers

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

The `loss` function measures the guessed answers against the known correct answers and measures how well or how badly it did.
The optimizer function to make another guess. Based on the loss function's result, it will try to minimize the loss.

```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
#Problem: consider the following 2 set of numbers
x_set = np.array([-1.0,0.0,1.0,2.0,3.0,4.0,], dtype=float)
y_set = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```
#traning the neural network
model.fit(x_set, y_set, epochs=500)
```

```
#use the model:model that has been trained to learn the relationship between X and Y.
#Y=3X+1

print(model.predict([10.0]))
```
